```python
import numpy as np
import pandas as pd
import os
import json
from sklearn.feature_extraction.text import TfidfVectorizer
import spacy
```

```python
files = []
subdir = tuple(os.walk("../imdb/imdb_movie_jsons/2010"))
for file_name in tuple(subdir)[0][-1]:
    with open(f"{subdir[0][0]}/{file_name}", 'r') as f:
        files.append(json.load(f))
```

```python
corpus = []
my_pos = set(["ADJ",])
nlp = spacy.load("en_core_web_sm")
nlp.max_length = 50_000_000
for file in files[::-1]:
    
    content = [review["content"] for review in file["reviews"]]
    for c in content:
        string = nlp(c)
        clean = []
        for word in string:
            if word.pos_ in my_pos:
                clean.append(str(word.lemma_))
        corpus.append(" ".join(clean))
    # corpus = clean
    break
```

```python
vect = TfidfVectorizer(stop_words="english", ngram_range=(1,1), use_idf=False, max_features=len(corpus)//3, sublinear_tf=True, norm="l1")
x = vect.fit_transform(corpus)
# vect.transform(corpus)
df = pd.DataFrame(x[0].T.toarray(), index=vect.get_feature_names_out(), columns=["TF-IDF"])
df = df.sort_values('TF-IDF', ascending=False)
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(df)
```
