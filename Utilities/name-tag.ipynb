```python
import json
import os
import re
```

```python
for movie_year in tuple(os.walk("./MergedData/merged_movie_jsons"))[0][1]:
    subdir = tuple(os.walk(f"./MergedData/merged_movie_jsons/{movie_year}"))
    for file_name in tuple(subdir)[0][-1]:
        with open(f"{subdir[0][0]}/{file_name}", 'r', encoding="utf-8") as f:
            file = json.load(f)
        new_file = {}
        name = re.search(r"(.*)-\d\d\d\d-\d\d-\d\d-(.*)\.json", file_name)
        if name:
            title = name.groups()[0]
            title = re.sub(r"--", r":", title)
            title = re.sub(r"(question_mark)", r"?", title)
            title = re.sub(r"(slash)", r"/", title)
            title = re.sub(r"(comma)", r",", title)
        new_file["title"] = title
        for key in file:
            new_file[key] = file[key]
        with open(f"{subdir[0][0]}/{file_name}", 'w', encoding="utf-8") as wr:
            json.dump(new_file, wr, indent=4)
```
